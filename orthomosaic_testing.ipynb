```python
import os
from pyodm import Node
```

```python
n = Node('localhost', 8000)
n.info()
```

```python
expected_paths = [f'data/RescueNet/train/train-org-img/{i}.jpg' for i in range(10778, 10787)]

# Filter out any missing files
image_paths = [path for path in expected_paths if os.path.exists(path)]
print(image_paths)
task = n.create_task(image_paths, { 'dsm': True })
task.wait_for_completion()
os.listdir(task.download_assets('results'))[0:9]
```
